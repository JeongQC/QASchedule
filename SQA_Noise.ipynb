```python
# ===============================
# SQA + Relative Noise + Tour Cost + Original Energy + TuRBO (Jupyter single-cell)
# ===============================

from dataclasses import dataclass, asdict
from types import SimpleNamespace
from pathlib import Path
import time, json, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import dimod
import openjij as oj
from pyqubo import Array, Placeholder, Constraint

# --------------------------------
# TSP instance + PyQUBO → BQM
# --------------------------------
def make_tsp_instance(n_customer: int, seed: int = 0):
    """
    Create TSP instance with n_customer + 1 points (including depot at (0,0)).
    Returns (df, distances):
      - df: DataFrame with columns ['x','y'] (index: 0..n_point-1)
      - distances: (n_point x n_point) float Euclidean distances
    """
    n_point = n_customer + 1
    np.random.seed(seed)
    df = pd.DataFrame({
        'x': np.random.randint(0, 100, n_point),
        'y': np.random.randint(0, 100, n_point),
    })
    # depot 고정
    df.loc[0, 'x'] = 0
    df.loc[0, 'y'] = 0

    coords = df[['x', 'y']].to_numpy(dtype=float)
    diff = coords[:, None, :] - coords[None, :, :]
    distances = np.sqrt((diff ** 2).sum(-1))
    return df, distances

def tsp_pyqubo(n_point, distances, penalty_scale=10.0):
    x = Array.create('c', shape=(n_point, n_point), vartype='BINARY')

    city_once = 0
    for i in range(n_point):
        city_once += Constraint((sum(x[i, k] for k in range(n_point)) - 1) ** 2, label=f"city_{i}")

    pos_once = 0
    for k in range(n_point):
        pos_once += Constraint((sum(x[i, k] for i in range(n_point)) - 1) ** 2, label=f"pos_{k}")

    dist_term = 0
    for k in range(n_point):
        kp1 = (k + 1) % n_point
        for i in range(n_point):
            xi = x[i, k]
            for j in range(n_point):
                dij = distances[i][j]
                if dij != 0:
                    dist_term += dij * xi * x[j, kp1]

    A = Placeholder("A")
    H = dist_term + A * (city_once + pos_once)
    model = H.compile()

    max_d = float(np.max(distances)) if n_point > 1 else 1.0
    feed = {"A": penalty_scale * max_d}
    bqm = model.to_bqm(feed_dict=feed)
    return model, bqm, feed

def load_bqm_for_experiment(n_customer: int, *, seed: int = 0, penalty_scale: float = 10.0):
    df, distances = make_tsp_instance(n_customer, seed=seed)
    _, bqm, _ = tsp_pyqubo(n_point=len(df), distances=distances, penalty_scale=penalty_scale)
    return bqm, df, distances

# --------------------------------
# Scale helpers (original energy)
# --------------------------------
def _bqm_max_abs_hJ(bqm: dimod.BinaryQuadraticModel):
    max_h = max((abs(v) for v in bqm.linear.values()), default=0.0)
    max_J = max((abs(w) for w in bqm.quadratic.values()), default=0.0)
    max_h = max_h if max_h > 0 else 1.0
    max_J = max_J if max_J > 0 else max_h
    return max_h, max_J

def compute_normalize_scale(bqm: dimod.BinaryQuadraticModel,
                            bias_range=(-2.0, 2.0),
                            quadratic_range=(-1.0, 1.0)) -> float:
    max_h, max_J = _bqm_max_abs_hJ(bqm)
    target_h = max(abs(bias_range[0]), abs(bias_range[1]))
    target_J = max(abs(quadratic_range[0]), abs(quadratic_range[1]))
    s_h = target_h / max_h if max_h > 0 else np.inf
    s_J = target_J / max_J if max_J > 0 else np.inf
    s = min(s_h, s_J)
    if not np.isfinite(s) or s <= 0:
        s = 1.0
    return float(s)

def normalize_bqm_inplace_and_get_scale(bqm: dimod.BinaryQuadraticModel,
                                        bias_range=(-2.0, 2.0),
                                        quadratic_range=(-1.0, 1.0)) -> float:
    s = compute_normalize_scale(bqm, bias_range, quadratic_range)
    if s != 1.0:
        bqm.scale(s)
    return s

# --------------------------------
# Relative noise → absolute
# --------------------------------
@dataclass
class NoiseConfigRel:
    sigma_h_rel: float = 0.0
    sigma_J_rel: float = 0.0
    systematic_h_rel: float = 0.0
    systematic_J_rel: float = 0.0
    frac_ghost: float = 0.0
    sigma_ghost_rel: float = 0.0
    schedule_s_sigma: float = 0.0
    beta_jitter_frac: float = 0.0
    gamma_jitter_frac: float = 0.0
    per_read: bool = False
    batches: int = 1
    seed: int | None = None

@dataclass
class NoiseConfig:
    sigma_h: float = 0.0
    sigma_J: float = 0.0
    systematic_h: float = 0.0
    systematic_J: float = 0.0
    frac_ghost: float = 0.0
    sigma_ghost: float = 0.0
    schedule_s_sigma: float = 0.0
    beta_jitter_frac: float = 0.0
    gamma_jitter_frac: float = 0.0
    per_read: bool = False
    batches: int = 1
    seed: int | None = None

def resolve_noise(bqm: dimod.BinaryQuadraticModel, rel: NoiseConfigRel) -> NoiseConfig:
    max_h, max_J = _bqm_max_abs_hJ(bqm)
    return NoiseConfig(
        sigma_h       = rel.sigma_h_rel * max_h,
        sigma_J       = rel.sigma_J_rel * max_J,
        systematic_h  = rel.systematic_h_rel * max_h,
        systematic_J  = rel.systematic_J_rel * max_J,
        frac_ghost    = rel.frac_ghost,
        sigma_ghost   = rel.sigma_ghost_rel * max_J,
        schedule_s_sigma = rel.schedule_s_sigma,
        beta_jitter_frac = rel.beta_jitter_frac,
        gamma_jitter_frac = rel.gamma_jitter_frac,
        per_read      = rel.per_read,
        batches       = rel.batches,
        seed          = rel.seed
    )

def _rng(seed):
    return np.random.default_rng(seed)

def perturb_bqm_once(bqm_in: dimod.BinaryQuadraticModel, cfg: NoiseConfig, rng) -> dimod.BinaryQuadraticModel:
    bqm = bqm_in.copy()
    if cfg.systematic_h != 0.0:
        for v in bqm.linear:
            bqm.linear[v] += cfg.systematic_h
    if cfg.systematic_J != 0.0:
        for (u, v) in bqm.quadratic:
            bqm.quadratic[(u, v)] += cfg.systematic_J
    if cfg.sigma_h > 0.0:
        for v in bqm.linear:
            bqm.linear[v] += rng.normal(0.0, cfg.sigma_h)
    if cfg.sigma_J > 0.0:
        for key in list(bqm.quadratic.keys()):
            bqm.quadratic[key] += rng.normal(0.0, cfg.sigma_J)
    if cfg.frac_ghost > 0.0 and cfg.sigma_ghost > 0.0:
        vars_list = list(bqm.variables)
        present = set(tuple(sorted(k)) for k in bqm.quadratic.keys())
        missing = []
        n = len(vars_list)
        for i in range(n):
            for j in range(i+1, n):
                pair = (vars_list[i], vars_list[j])
                if pair not in present:
                    missing.append(pair)
        if missing:
            m = int(np.ceil(cfg.frac_ghost * len(missing)))
            idx = rng.choice(len(missing), size=m, replace=False)
            for k in idx:
                u, v = missing[k]
                bqm.add_interaction(u, v, rng.normal(0.0, cfg.sigma_ghost))
    return bqm

def jitter_schedule_s(schedule: list[tuple[float,int]], cfg: NoiseConfig, rng) -> list[tuple[float,int]]:
    if cfg.schedule_s_sigma <= 0.0:
        return schedule
    s = np.array([s for (s, step) in schedule], dtype=float)
    step = [st for (s, st) in schedule]
    s = s + rng.normal(0.0, cfg.schedule_s_sigma, size=s.shape)
    s = np.clip(s, 0.0, 1.0)
    s = np.maximum.accumulate(s)
    return [(float(si), int(st)) for si, st in zip(s, step)]

# --------------------------------
# Schedule builder (s-domain)
# --------------------------------
def make_openjij_schedule(u_rs: np.ndarray):
    return [(float(s), 1) for s in u_rs]

def make_fourier_u_s(thetas, n_point=201, enforce_monotone=True):
    thetas = np.asarray(thetas, dtype=float)
    s = np.linspace(0.0, 1.0, int(n_point))
    u = s.copy()
    for j, theta_j in enumerate(thetas, start=1):
        u += theta_j * np.sin(j * np.pi * s)
    u = np.clip(u, 0.0, 1.0)
    if enforce_monotone:
        u = np.maximum.accumulate(u)
    return u

# --------------------------------
# Decoders & tour cost helpers
# --------------------------------
def _decode_X(sample_vec, variables, n_point, var_name_fmt="c[{i}][{j}]"):
    X = np.zeros((n_point, n_point), dtype=int)
    idx = {v: k for k, v in enumerate(variables)}
    for i in range(n_point):
        for j in range(n_point):
            k = idx.get(var_name_fmt.format(i=i, j=j), None)
            if k is not None:
                X[i, j] = int(sample_vec[k])
    return X

def _tour_cost_from_X(X, distances):
    n_point = X.shape[0]
    cost = 0.0
    for k in range(n_point):
        i = int(np.argmax(X[:, k]))
        j = int(np.argmax(X[:, (k + 1) % n_point]))
        cost += float(distances[i, j])
    return float(cost)

def _eval_tour_costs(samples, variables, n_point, distances, var_name_fmt="c[{i}][{j}]"):
    feas = []
    costs = []
    for s in samples:
        X = _decode_X(s, variables, n_point, var_name_fmt)
        ok = (X.sum(axis=0) == 1).all() and (X.sum(axis=1) == 1).all()
        feas.append(ok)
        if ok:
            costs.append(_tour_cost_from_X(X, distances))
        else:
            costs.append(np.inf)
    feas = np.array(feas, dtype=bool)
    costs = np.array(costs, dtype=float)
    best_feas_cost = float(costs[feas].min()) if feas.any() else np.inf
    return feas, costs, best_feas_cost

# --------------------------------
# SQA runners (energy original)
# --------------------------------
def sqa_sample_with_schedule(
    bqm,
    schedule,
    *,
    num_reads: int = 1000,
    trotter: int = 8,
    beta_max: float = 5.0,
    gamma_max: float = 2.0,
    energy_scale: float = 1.0,
):
    num_sweeps = len(schedule)
    sampler = oj.SQASampler()
    t0 = time.time()
    try:
        ss = sampler.sample(
            bqm, num_reads=num_reads, schedule=schedule, num_sweeps=num_sweeps,
            trotter=trotter, beta=beta_max, gamma=gamma_max
        )
    except TypeError:
        ss = sampler.sample(
            bqm, num_reads=num_reads, schedule=schedule, num_sweeps=num_sweeps,
            beta=beta_max, gamma=gamma_max
        )
    wall = time.time() - t0

    energies_norm = ss.record.energy.astype(float)
    energies_orig = energies_norm / max(energy_scale, 1e-12)
    variables = list(ss.variables)
    samples = ss.record.sample

    e_min_norm = float(energies_norm.min())
    p_succ_min = float((energies_norm == e_min_norm).mean())
    e_min_orig = float(e_min_norm / max(energy_scale, 1e-12))

    info = dict(
        best_energy=e_min_norm,
        best_energy_original=e_min_orig,
        p_succ_min=p_succ_min,
        num_reads=int(num_reads),
        trotter=int(trotter),
        wall_sec=wall,
        energies=energies_norm,
        energies_original=energies_orig,
        variables=variables,
        samples=samples,
        energy_scale=float(energy_scale),
    )
    return ss, info

def sqa_sample_with_schedule_noisy(
    bqm,
    schedule,
    *,
    num_reads: int = 1000,
    trotter: int = 8,
    beta_max: float = 5.0,
    gamma_max: float = 2.0,
    noise: NoiseConfig | None = None,
    energy_scale: float = 1.0,
):
    if (noise is None) or (noise.batches <= 1 and not noise.per_read):
        return sqa_sample_with_schedule(
            bqm, schedule,
            num_reads=num_reads, trotter=trotter,
            beta_max=beta_max, gamma_max=gamma_max,
            energy_scale=energy_scale
        )

    rng = _rng(noise.seed)
    batches = max(1, int(noise.batches))
    reads_per = [num_reads // batches] * batches
    for i in range(num_reads % batches):
        reads_per[i] += 1

    energies_all, samples_all = [], []
    variables_ref = None
    t0 = time.time()

    for bi, r in enumerate(reads_per):
        bqm_batch = perturb_bqm_once(bqm, noise, rng) if (noise.per_read or bi == 0) else bqm
        sch_batch = jitter_schedule_s(schedule, noise, rng)
        beta_b = beta_max * (1.0 + (rng.normal(0.0, noise.beta_jitter_frac) if noise.beta_jitter_frac>0 else 0.0))
        gamma_b = gamma_max * (1.0 + (rng.normal(0.0, noise.gamma_jitter_frac) if noise.gamma_jitter_frac>0 else 0.0))
        try:
            ss = oj.SQASampler().sample(
                bqm_batch, num_reads=r, schedule=sch_batch, num_sweeps=len(sch_batch),
                trotter=trotter, beta=beta_b, gamma=gamma_b
            )
        except TypeError:
            ss = oj.SQASampler().sample(
                bqm_batch, num_reads=r, schedule=sch_batch, num_sweeps=len(sch_batch),
                beta=beta_b, gamma=gamma_b
            )
        if variables_ref is None:
            variables_ref = list(ss.variables)
        energies_all.append(ss.record.energy.astype(float))
        samples_all.append(ss.record.sample)

    wall = time.time() - t0
    energies_norm = np.concatenate(energies_all, axis=0) if energies_all else np.empty((0,))
    samples = np.concatenate(samples_all, axis=0) if samples_all else np.empty((0, len(variables_ref) if variables_ref else 0))

    e_min_norm = float(energies_norm.min()) if energies_norm.size else np.inf
    p_succ_min = float((energies_norm == e_min_norm).mean()) if energies_norm.size else 0.0
    energies_orig = energies_norm / max(energy_scale, 1e-12)
    e_min_orig = float(e_min_norm / max(energy_scale, 1e-12)) if np.isfinite(e_min_norm) else np.inf

    info = dict(
        best_energy=e_min_norm,
        best_energy_original=e_min_orig,
        p_succ_min=p_succ_min,
        num_reads=int(num_reads),
        trotter=int(trotter),
        wall_sec=wall,
        energies=energies_norm,
        energies_original=energies_orig,
        variables=variables_ref,
        samples=samples,
        energy_scale=float(energy_scale),
    )
    return None, info

# --------------------------------
# Evaluator (θ → schedule → SQA) with tour cost + snapshots
# --------------------------------
def evaluate_schedule_with_sqa_s(
    bqm,
    thetas,
    *,
    n_point,
    distances,                       # ← 거리 행렬
    num_points=200,
    num_reads=1000,
    trotter=8,
    beta_max=5.0,
    gamma_max=2.0,
    enforce_monotone=True,
    var_name_fmt="c[{i}][{j}]",
    noise_rel: NoiseConfigRel | None = None,
    energy_scale: float | None = None,
    auto_compute_scale: bool = False,
    bias_range=(-2.0,2.0),
    quadratic_range=(-1.0,1.0)
):
    internal_scale = 1.0
    if energy_scale is not None:
        internal_scale = float(energy_scale)
    elif auto_compute_scale:
        internal_scale = normalize_bqm_inplace_and_get_scale(bqm, bias_range, quadratic_range)

    u = make_fourier_u_s(thetas, n_point=num_points, enforce_monotone=enforce_monotone)
    schedule = make_openjij_schedule(u)

    resolved = resolve_noise(bqm, noise_rel) if noise_rel is not None else None
    if resolved is None:
        _, info = sqa_sample_with_schedule(
            bqm, schedule,
            num_reads=num_reads, trotter=trotter,
            beta_max=beta_max, gamma_max=gamma_max,
            energy_scale=internal_scale
        )
    else:
        _, info = sqa_sample_with_schedule_noisy(
            bqm, schedule,
            num_reads=num_reads, trotter=trotter,
            beta_max=beta_max, gamma_max=gamma_max,
            noise=resolved,
            energy_scale=internal_scale
        )

    samples   = info["samples"]
    variables = info["variables"] if info["variables"] is not None else []

    # 1) feasibility success_prob (행/열 합 = 1)
    feas_flags = []
    if len(samples):
        for s in samples:
            X = _decode_X(s, variables, n_point, var_name_fmt)
            feas_flags.append((X.sum(axis=0) == 1).all() and (X.sum(axis=1) == 1).all())
    feas_flags = np.array(feas_flags, dtype=bool) if len(samples) else np.array([], bool)
    info["success_prob"] = float(feas_flags.mean()) if len(feas_flags) else 0.0

    # 2) tour cost (feasible만)
    if len(samples):
        feas2, tour_costs, best_tour = _eval_tour_costs(samples, variables, n_point, distances, var_name_fmt)
    else:
        feas2 = np.array([], bool); tour_costs = np.array([], float); best_tour = np.inf
    info["tour_costs"] = tour_costs
    info["tour_cost_min"] = float(best_tour)
    info["tour_cost_mean_feas"] = float(np.mean(tour_costs[feas2])) if feas2.any() else np.inf

    # 3) snapshots
    info["noise_snapshot_rel"] = asdict(noise_rel) if noise_rel is not None else None
    info["noise_snapshot_abs"] = asdict(resolved)  if noise_rel is not None else None
    return info["best_energy"], info

# --------------------------------
# TuRBO (shape-only, s-domain) — logs: energy & tour
# --------------------------------
def turbo_optimize_schedule_shape_only_sqa_s(
    bqm,
    *,
    n_point,
    distances,                 # ← 추가: 도메인 비용 계산용
    num_points=200,
    num_reads=1000,
    trotter=8,
    beta_max=5.0,
    gamma_max=2.0,
    n_harmonics=4,
    n_calls=40,
    n_rep=3,
    random_state=42,
    enforce_monotone=True,
    init_tr_frac=0.5,
    min_tr_frac=0.05,
    grow=1.6,
    shrink=0.5,
    succ_tol=3,
    fail_tol=3,
    progress=True,
    progress_every=1,
    noise_rel: NoiseConfigRel | None = None,
    energy_scale: float | None = None,
    auto_compute_scale: bool = False,
    bias_range=(-2.0,2.0),
    quadratic_range=(-1.0,1.0)
):
    rng = np.random.default_rng(random_state)
    lb = np.array([-1.0/j for j in range(1, n_harmonics+1)], dtype=float)
    ub = np.array([ 1.0/j for j in range(1, n_harmonics+1)], dtype=float)
    span = ub - lb

    x_best = np.zeros(n_harmonics, dtype=float)
    f_best = np.inf; f_best_orig = np.inf
    best_info = None

    # 초기 평가
    vals = []; vals_orig=[]; infos0=[]; tours0=[]
    for _ in range(n_rep):
        f, info = evaluate_schedule_with_sqa_s(
            bqm, x_best,
            n_point=n_point,
            distances=distances,
            num_points=num_points, num_reads=num_reads,
            trotter=trotter, beta_max=beta_max, gamma_max=gamma_max,
            enforce_monotone=enforce_monotone,
            noise_rel=noise_rel,
            energy_scale=energy_scale,
            auto_compute_scale=auto_compute_scale,
            bias_range=bias_range,
            quadratic_range=quadratic_range
        )
        vals.append(f)
        vals_orig.append(info["best_energy_original"])
        tours0.append(info["tour_cost_min"])
        infos0.append(info)
    f_cur = float(np.mean(vals)); f_cur_orig = float(np.mean(vals_orig))
    tour_cur = float(np.min(tours0))
    f_best = f_cur; f_best_orig = f_cur_orig
    tour_best = tour_cur
    best_info = infos0[int(np.argmin(vals))]
    x_cur = x_best.copy()

    L = init_tr_frac * span
    succ_streak = 0; fail_streak = 0

    eval_log = []; cand_log = []

    def project(x): return np.minimum(ub, np.maximum(lb, x))
    def _flatten_snap(info_dict: dict, prefix: str):
        snap = info_dict.get(prefix)
        if snap is None: return {}
        return {f"{prefix}.{k}": v for k, v in snap.items()}

    for it in range(1, n_calls+1):
        delta = (rng.uniform(-0.5, 0.5, size=n_harmonics)) * L
        x_try = project(x_cur + delta)

        vals = []; vals_o=[]; infos=[]; tours=[]
        for rep_idx in range(n_rep):
            f_val, info = evaluate_schedule_with_sqa_s(
                bqm, x_try,
                n_point=n_point,
                distances=distances,
                num_points=num_points, num_reads=num_reads,
                trotter=trotter, beta_max=beta_max, gamma_max=gamma_max,
                enforce_monotone=enforce_monotone,
                noise_rel=noise_rel,
                energy_scale=energy_scale,
                auto_compute_scale=auto_compute_scale,
                bias_range=bias_range,
                quadratic_range=quadratic_range
            )
            vals.append(f_val); vals_o.append(info["best_energy_original"]); infos.append(info)
            tours.append(info["tour_cost_min"])

            row = {
                "iter": it, "rep": rep_idx,
                "f_val": float(f_val),
                "f_val_original": float(info["best_energy_original"]),
                "tour_cost_min": float(info["tour_cost_min"]),
                "tour_cost_mean_feas": float(info["tour_cost_mean_feas"]),
                "success_prob": float(info["success_prob"]),
                **{f"theta_{j+1}": float(x_try[j]) for j in range(n_harmonics)}
            }
            row.update(_flatten_snap(info, "noise_snapshot_rel"))
            row.update(_flatten_snap(info, "noise_snapshot_abs"))
            eval_log.append(row)

        f_try = float(np.mean(vals)); f_try_orig = float(np.mean(vals_o))
        tour_try = float(np.min(tours))
        sp_try = float(np.mean([inf["success_prob"] for inf in infos]))
        info_agg = infos[int(np.argmin(vals))]

        improved = (f_try < f_cur - 1e-12)
        if improved:
            x_cur = x_try; f_cur = f_try; f_cur_orig = f_try_orig; tour_cur = tour_try
            succ_streak += 1; fail_streak = 0
            if f_try < f_best:
                f_best = f_try; f_best_orig = f_try_orig
                tour_best = min(tour_best, tour_try)
                x_best = x_try.copy(); best_info = info_agg
        else:
            fail_streak += 1; succ_streak = 0

        if succ_streak >= succ_tol:
            L = np.minimum(span, L * grow);  succ_streak = 0
        elif fail_streak >= fail_tol:
            L = np.maximum(min_tr_frac * span, L * shrink); fail_streak = 0

        tr_frac = float(np.mean(L / span))
        row_iter = {
            "iter": it,
            "obj_val": f_try,
            "obj_val_original": f_try_orig,
            "obj_tour": tour_try,
            "incumbent_best": f_best,
            "incumbent_best_original": f_best_orig,
            "incumbent_tour": tour_best,
            "success_prob": sp_try,
            "tr_frac": tr_frac,
            **{f"theta_{j+1}": float(x_cur[j]) for j in range(n_harmonics)}
        }
        row_iter.update(_flatten_snap(info_agg, "noise_snapshot_rel"))
        row_iter.update(_flatten_snap(info_agg, "noise_snapshot_abs"))
        cand_log.append(row_iter)

        if progress and (it % progress_every == 0):
            print(f"[TuRBO] iter={it:03d}  obj={f_try:.6g} (orig {f_try_orig:.6g})  "
                  f"tour={tour_try:.3f}  best={f_best:.6g}  tour*={tour_best:.3f}  "
                  f"P≈{sp_try:.3f}  TR≈{tr_frac:.3f}")

    df_eval = pd.DataFrame(eval_log)
    df_cand = pd.DataFrame(cand_log)
    # turbo_optimize_schedule_shape_only_sqa_s 끝부분
    best = SimpleNamespace(
        best_energy=f_best,
        best_energy_original=f_best_orig,
        params={"thetas": x_best.tolist()},
        schedule=make_openjij_schedule(
            make_fourier_u_s(x_best, n_point=num_points, enforce_monotone=enforce_monotone)
        ),
        schedule_u=make_fourier_u_s(x_best, n_point=num_points, enforce_monotone=enforce_monotone).tolist(),
        tts=None,
        success_prob=best_info["success_prob"] if best_info else np.nan
    )
    return best, df_eval, df_cand


# --------------------------------
# Repeated experiments wrapper
# --------------------------------
def run_experiments(
    Ns=(3,4,5,6,7,8,9),
    *,
    seed_inst: int = 0,
    penalty_scale: float = 5.0,
    noise_rel: NoiseConfigRel | None = None,
    # SQA/TuRBO hyperparams
    num_points: int = 200,
    num_reads: int = 250,
    trotter: int = 8,
    beta_max: float = 5.0,
    gamma_max: float = 2.0,
    n_harmonics: int = 4,
    n_calls: int = 40,
    n_rep: int = 3,
    random_state: int = 42,
    enforce_monotone: bool = True,
    # energy scaling strategy
    normalize_strategy: str = "none",  # "none" | "auto" | "external"
    external_scale: float | None = None,
    bias_range = (-2.0, 2.0),
    quadratic_range = (-1.0, 1.0),
    # output/plot
    outdir: str = "./results_sqa_turbo",
    do_plot: bool = True
):
    """
    normalize_strategy:
      - "none"     : normalize 안 함 (tour-cost 그대로). energy_scale=None
      - "auto"     : 내부에서 scale 계산/적용. energy_original 복원됨.
      - "external" : 사용자가 미리 bqm을 scale했음. external_scale 필수(또는 여기서 계산/적용).
    """
    outdir = Path(outdir); outdir.mkdir(parents=True, exist_ok=True)
    rows_summary = []

    for N in Ns:
        print(f"\n====== N={N} (n_customer) ======")
        bqm, df_inst, D = load_bqm_for_experiment(n_customer=N, seed=seed_inst, penalty_scale=penalty_scale)
        n_point = len(df_inst)

        energy_scale = None
        auto_compute_scale = False
        if normalize_strategy == "none":
            pass
        elif normalize_strategy == "auto":
            auto_compute_scale = True
        elif normalize_strategy == "external":
            if external_scale is None:
                s = normalize_bqm_inplace_and_get_scale(bqm, bias_range, quadratic_range)
                energy_scale = s
            else:
                energy_scale = float(external_scale)
        else:
            raise ValueError("normalize_strategy must be in {'none','auto','external'}")

        best, df_eval, df_cand = turbo_optimize_schedule_shape_only_sqa_s(
            bqm,
            n_point=n_point,
            distances=D,                         # ← 거리 행렬 전달
            num_points=num_points,
            num_reads=num_reads,
            trotter=trotter,
            beta_max=beta_max,
            gamma_max=gamma_max,
            n_harmonics=n_harmonics,
            n_calls=n_calls,
            n_rep=n_rep,
            random_state=random_state,
            enforce_monotone=enforce_monotone,
            init_tr_frac=0.5, min_tr_frac=0.05,
            grow=1.6, shrink=0.5, succ_tol=3, fail_tol=3,
            progress=True, progress_every=max(1, n_calls//8),
            noise_rel=noise_rel,
            energy_scale=energy_scale,
            auto_compute_scale=auto_compute_scale,
            bias_range=bias_range,
            quadratic_range=quadratic_range
        )

        # Save per-N
        tag = f"N{N}"
        df_eval.to_csv(outdir / f"eval_{tag}.csv", index=False)
        df_cand.to_csv(outdir / f"cand_{tag}.csv", index=False)
        with open(outdir / f"best_{tag}.json", "w", encoding="utf-8") as f:
            json.dump({
                "N": int(N),
                "best_energy_norm": float(best.best_energy),
                "best_energy_original": float(best.best_energy_original),
                "thetas": best.params["thetas"],
            }, f, ensure_ascii=False, indent=2)

        inc_tour = float(df_cand["incumbent_tour"].dropna().iloc[-1]) if "incumbent_tour" in df_cand.columns else np.nan
        # run_experiments 내부 rows_summary.append
        rows_summary.append({
            "N": int(N),
            "n_point": int(n_point),
            "best_norm": float(best.best_energy),
            "best_orig": float(best.best_energy_original),
            "best_tour": float(inc_tour),
            "success_prob": float(best.success_prob),
            "schedule_u": json.dumps(best.schedule_u),   # JSON string으로 저장
            "reads": int(num_reads),
            "trotter": int(trotter),
            "beta_max": float(beta_max),
            "gamma_max": float(gamma_max),
            "n_calls": int(n_calls),
            "n_rep": int(n_rep),
            "normalize_strategy": normalize_strategy,
            "energy_scale": (None if energy_scale is None else float(energy_scale)),
        })


        # (optional) quick plot per N — tour(cost)
        if do_plot:
            plt.figure()
            plt.plot(df_cand["iter"], df_cand["obj_tour"])
            plt.xlabel("Iteration"); plt.ylabel("Tour cost (domain)"); plt.title(f"Tour cost per-iter (N={N})")
            plt.show()

    df_sum = pd.DataFrame(rows_summary).sort_values("N")
    df_sum.to_csv(outdir / "summary_all.csv", index=False)
    print(f"\nSaved all logs to: {outdir.resolve()}")
    return df_sum

# ===========================
# USER CONFIG (edit below)
# ===========================
noise_rel = NoiseConfigRel(
    sigma_h_rel=0.05, sigma_J_rel=0.02,
    frac_ghost=0.02, sigma_ghost_rel=0.01,
    schedule_s_sigma=0.001,
    beta_jitter_frac=0.01, gamma_jitter_frac=0.02,
    per_read=True, batches=4, seed=0
)

# 예시 실행: 여러 N 반복 (투어코스트 스케일 유지)
df_summary = run_experiments(
    Ns=(1,2,3,4,5,6,7,8,9),           # ← 원하는 N 범위로 수정
    seed_inst=0,
    penalty_scale=5.0,
    noise_rel=noise_rel,
    num_points=200,
    num_reads=2000,
    trotter=8,
    beta_max=5.0,
    gamma_max=2.0,
    n_harmonics=4,
    n_calls=100,           # 데모 20; 본실험은 40+ 권장
    n_rep=3,
    random_state=42,
    normalize_strategy="none",   # "none" | "auto" | "external"
    external_scale=None,
    outdir="./results_sqa_turbo",
    do_plot=True
)

display(df_summary)
```
